## Import Semua Packages/Library yang Digunakan

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import streamlit as st
```

## Data Wrangling

### Gathering Data

```python
hourly_data = pd.read_csv('data\hour.csv')
hourly_data.head()
```

### Assessing Data

```python
# Menampilkan lima baris pertama data
print(hourly_data.head())

# Melihat informasi umum tentang data
print(hourly_data.info())

# Melihat statistik deskriptif
print(hourly_data.describe())
```

### Cleaning Data

```python
# Mengganti nilai yang hilang dengan nilai rata-rata dari kolom-kolom numerik
hourly_data.fillna(hourly_data.mean(numeric_only=True), inplace=True)
```

## Exploratory Data Analysis (EDA)

### Explore ...

```python

# Menampilkan informasi data
st.write("Informasi Data:")
st.write(hourly_data.info())

# Menampilkan beberapa baris pertama data
st.write("\nBeberapa Baris Pertama Data:")
st.write(hourly_data.head())

# Statistik deskriptif
st.write("\nStatistik Deskriptif:")
st.write(hourly_data.describe())


# Visualisasi korelasi antar variabel numerik
plt.figure(figsize=(10, 8))
sns.heatmap(hourly_data.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Korelasi Antara Variabel Numerik')
plt.show()
```

```python
# Histogram
plt.figure(figsize=(12, 6))
sns.histplot(hourly_data['cnt'], bins=30, kde=True, color='skyblue')
plt.title('Distribusi Jumlah Sewa Sepeda')
plt.xlabel('Jumlah Sewa Sepeda')
plt.ylabel('Frekuensi')
plt.show()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
### Pertanyaan 1: Hubungan antara faktor lingkungan dan musiman dengan jumlah sewa sepeda secara keseluruhan
plt.figure(figsize=(10, 6))
sns.scatterplot(data=hourly_data, x='temp', y='cnt', hue='season')
plt.title('Hubungan antara Suhu dan Jumlah Sewa Sepeda')
plt.xlabel('Suhu (Celsius)')
plt.ylabel('Jumlah Sewa Sepeda')
plt.legend(title='Musim')
plt.show()
```

### Pertanyaan 2:

```python
### Pertanyaan 2: Mendeteksi anomali atau kejadian penting di kota Washington D.C. yang berdampak pada jumlah sewa sepeda

# Box Plot antara Hari Libur dan Jumlah Sewa Sepeda
plt.figure(figsize=(10, 6))
sns.boxplot(data=hourly_data, x='holiday', y='cnt')
plt.title('Pengaruh Hari Libur terhadap Jumlah Sewa Sepeda')
plt.xlabel('Hari Libur')
plt.ylabel('Jumlah Sewa Sepeda')
plt.xticks([0, 1], ['Bukan Hari Libur', 'Hari Libur'])
plt.show()
```

## Conclusion

1. Dari hasil analisis yang dilakukan, dapat disimpulkan bahwa terdapat hubungan antara faktor lingkungan dan musiman dengan jumlah sewa sepeda secara keseluruhan:
- Dari scatter plot antara suhu dan jumlah sewa sepeda, terlihat bahwa ada hubungan positif antara suhu dan jumlah sewa sepeda, di mana jumlah sewa sepeda cenderung meningkat saat suhu meningkat.
- Box plot antara musim dan jumlah sewa sepeda menunjukkan bahwa musim juga memiliki pengaruh signifikan terhadap jumlah sewa sepeda, dengan musim musim panas dan musim gugur cenderung memiliki jumlah sewa sepeda yang lebih tinggi daripada musim lainnya.

2. Berdasarkan analisis terhadap kejadian penting di kota Washington D.C. dan dampaknya terhadap jumlah sewa sepeda, diperoleh beberapa kesimpulan:
- Box plot antara hari libur dan jumlah sewa sepeda menunjukkan bahwa ada perbedaan yang signifikan dalam jumlah sewa sepeda antara hari libur dan bukan hari libur, di mana jumlah sewa sepeda cenderung lebih tinggi saat bukan hari libur.
- Analisis perubahan jumlah sewa sepeda selama rentang waktu tertentu menunjukkan adanya fluktuasi yang signifikan, dengan beberapa periode tertentu mungkin mengalami peningkatan atau penurunan jumlah sewa sepeda yang tidak biasa. Hal ini menunjukkan bahwa kejadian penting atau anomali tertentu mungkin memiliki dampak langsung terhadap jumlah sewa sepeda di kota tersebut.
